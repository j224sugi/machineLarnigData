```python
import pandas as pd
import os

encoding="cp932"
def getBlocks(textFile):
    with open(textFile,"r") as f:
        textString=f.read()
    blocks=textString.split("\n")
    return blocks

def getProjectInfo(block):
    projectInfo=block.split(",")
    projectName=projectInfo[0]
    return projectName

def getProjectSmellDF(projectName,SmellName):
    targetFile=os.path.join("joinCKSPOON_4",projectName,SmellName+".csv")
    projectSmell=pd.read_csv(targetFile,encoding=encoding,encoding_errors="replace")
    return projectSmell

```

```python
SmellsName=["GodClass","RefusedBequest","FeatureEnvy","LongMethod"]
blocks=getBlocks("version.txt")
saveFolder=os.path.join(os.getcwd(),"dataset")
if not os.path.isdir(saveFolder):
    os.makedirs(saveFolder)

for smellName in SmellsName:
    concatDF=pd.DataFrame()
    for block in blocks:
        projectName=getProjectInfo(block)
        df=getProjectSmellDF(projectName,smellName)
        concatDF=pd.concat([concatDF,df])
        
    concatDF.to_csv(os.path.join(saveFolder,smellName+".csv"),index=False)
    
    
        
```
